```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://momq.co.kr/') 
```

```python
def review(df):
    if ('배송' in df) | ('박스' in df) | ('반품' in df) | ('상자' in df) | ('오배송' in df) | ('착오발송' in df) | ('상자' in df):
        a = '배송문제'
    elif ('싸' in df) | ('가격' in df) | ('이벤트' in df) | ('행사' in df):
        a = '가격문제'
    elif ('제조' in df):
        a = '제조기한'
    else:
        a = '기타'
    return a

review_diaper['리뷰_분류'] = review_diaper['리뷰'].apply(review)
```
